```python
## Task 1: Mark all launch sites on a map

# Define the launch site locations
launch_sites = [
    {"name": "CCAFS LC-40", "lat": 28.562302, "long": -80.577356},
    {"name": "CCAFS SLC-40", "lat": 28.563197, "long": -80.576820},
    {"name": "KSC LC-39A", "lat": 28.573255, "long": -80.646895},
    {"name": "VAFB SLC-4E", "lat": 34.632834, "long": -120.610745}
]

# Calculate the average latitude and longitude for centering the map
avg_lat = sum([site["lat"] for site in launch_sites]) / len(launch_sites)
avg_long = sum([site["long"] for site in launch_sites]) / len(launch_sites)

# Create a folium map centered at the average location
launch_map = folium.Map(location=[avg_lat, avg_long], zoom_start=5)

# Add markers for each launch site
for site in launch_sites:
    folium.Marker(
        location=[site["lat"], site["long"]],
        popup=site["name"],
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(launch_map)

# Display the map
launch_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map


launch_data = [
    {"site": "CCAFS LC-40", "lat": 28.562302, "long": -80.577356, "outcome": "Success"},
    {"site": "CCAFS LC-40", "lat": 28.562302, "long": -80.577356, "outcome": "Failure"},
    {"site": "CCAFS SLC-40", "lat": 28.563197, "long": -80.576820, "outcome": "Success"},
    {"site": "KSC LC-39A", "lat": 28.573255, "long": -80.646895, "outcome": "Failure"},
    {"site": "VAFB SLC-4E", "lat": 34.632834, "long": -120.610745, "outcome": "Success"},
    {"site": "VAFB SLC-4E", "lat": 34.632834, "long": -120.610745, "outcome": "Failure"}
]

# Define marker colors based on success or failure
def get_marker_color(outcome):
    return "green" if outcome == "Success" else "red"

# Add markers for success and failure launches
for launch in launch_data:
    folium.Marker(
        location=[launch["lat"], launch["long"]],
        popup=f"Site: {launch['site']}<br>Outcome: {launch['outcome']}",
        icon=folium.Icon(color=get_marker_color(launch["outcome"]), icon="ok" if launch["outcome"] == "Success" else "remove")
    ).add_to(site_map)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities

import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Distance Calculation Function
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance


# Coordinates for Launch Site and Coastline
lat_lon_launch_site = [28.56318, -80.57683]
lat_lon_coastline = [28.56317, -80.56796]

# Extract Latitude and Longitude
launch_site_lat = lat_lon_launch_site[0]
launch_site_lon = lat_lon_launch_site[1]
coastline_lat = lat_lon_coastline[0]
coastline_lon = lat_lon_coastline[1]

# Calculate Distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance to coastline: {distance_coastline:.2f} KM")


# Folium Map Setup
site_map = folium.Map(location=lat_lon_launch_site, zoom_start=15)

# Add Distance Marker at the midpoint between launch site and coastline
midpoint = [(launch_site_lat + coastline_lat) / 2, (launch_site_lon + coastline_lon) / 2]
distance_marker = folium.Marker(
    midpoint,
    icon=DivIcon(
        icon_size=(100, 30),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_coastline)
    )
)
site_map.add_child(distance_marker)

# Add Polyline between Launch Site and Coastline
coordinates = [lat_lon_launch_site, lat_lon_coastline]
lines = folium.PolyLine(
    locations=coordinates,
    weight=2,
    color='blue',
    dash_array='5,5'
)
site_map.add_child(lines)

# Add Markers for Launch Site and Coastline
folium.Marker(
    lat_lon_launch_site,
    popup='Launch Site',
    icon=folium.Icon(color='green', icon='rocket')
).add_to(site_map)

folium.Marker(
    lat_lon_coastline,
    popup='Coastline',
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(site_map)

# Save Map
site_map.save('launch_site_to_coastline.html')

# Display the map (Jupyter Notebook only)
site_map

```
